```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

### Soal Nomor 3
#### Webscrapping from the link given

```python
# Perolehan medali SEAgames 2017

url = requests.get("https://www2.2019seagames.com/countries/")
data = BeautifulSoup(url.content, "html.parser")

# All necessary data are in <em>

a = []

for i in data.find_all("em"):
    a.append(i.text)

a
```

```python
# Getting the country list

country = a[::6]

country = country[0:11]

country
```

```python
# Getting the gold medal list

goldS= a[2::6]

goldS
```

```python
# Gettting the number of the gold medal list

goldS1 = []
gold = []

for i in range(len(goldS)):
    temp = goldS[i].split(" ")
    goldS1.append(temp[1])
    
for j in range(len(goldS1)):
    temp = int(goldS1[j])
    gold.append(temp)
    
gold = gold[:11]

gold
```

```python
# Perolehan Medali SEAgames 2019

url1 = requests.get("https://rs.2019seagames.com/RS2019/mobiapp/MedalTally")
data1 = BeautifulSoup(url1.content, "html.parser")

# All necessary datas are in <small>
a1 = []

for i in data1.find_all("small"):
    a1.append(i.text)

a1
```

```python
a2 = a1[5:]

a2
```

```python
# Getting the list of the country
country1 = a2[::5]

country1
```

```python
# Getting the number of gold medal list
gold2019 = a2[1::5]

gold2019
```

### Merging country and number of medals achievement 

```python

all = list(zip(country, gold))
all.sort()

all1 = list(zip(country1, gold2019))
all1.sort()

all
```

```python
all1
```

### 2017 & 2019 Countries

```python
country_all = []

for i in range(len(all1)):
    temp = all1[i]
    country_all.append(temp[0])

country_all
```

### Merging gold medal in 2017 and 2019

```python
gold_2017 = []

for i in range(len(all)):
    temp = all[i]
    gold_2017.append(temp[1])

gold_2017
```

```python
gold_2019 = []

for i in range(len(all1)):
    temp = all1[i]
    gold_2019.append(int(temp[1]))

gold_2019
```

### Visualizing the data 

```python
plt.figure(figsize=(15,7))
plt.plot(country_all, gold_2017, "r-o")
plt.plot(country_all, gold_2019, "g-o")
plt.plot(country_all[gold_2017.index(max(gold_2017))], max(gold_2017), '*', markersize = 20, markerfacecolor = "gold")
plt.plot(country_all[gold_2019.index(max(gold_2019))], max(gold_2019), '*', markersize = 20, markerfacecolor = "yellow")
plt.title('GOLD MEDAL TALLY SEA GAMES 2017 & 2019', fontsize = 30, pad = 60)
plt.xlabel('COUNTRY', fontsize = 20)
plt.ylabel('JUMLAH MEDALI', fontsize = 20)
plt.xticks(rotation = 25)
plt.legend(['Gold 2017', 'Gold 2019', 'Champion 2017', 'Champion 2019'])
plt.grid(True)
plt.show
```

```python
plt.figure(figsize = (15,7))
plt.subplot(1, 2, 1)
plt.pie(gold_2017, labels = country_all, autopct='%1.1f%%')
plt.title('%GOLD SEA GAMES 2017', fontsize = 20)

plt.subplot(1, 2, 2)
plt.pie(gold_2019, labels = country_all, autopct='%1.1f%%')
plt.title('%GOLD SEA GAMES 2019', fontsize = 20)
plt.show()
```
